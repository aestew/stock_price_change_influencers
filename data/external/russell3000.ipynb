### Final Dataframe for YFinance

```python
# Getting adj close and delta, volume, dividends, stock

# Will pull first 60 days
start_date = "2025-01-20"
end_date = "2025-03-22" # Need to specify March 22nd to pull previous day

companies = r3000_tickers
tickers = yf.Tickers(companies)
tickers_hist = tickers.history(period='max',start=start_date, end=end_date,interval='1d', auto_adjust=False)

adj_close = tickers_hist['Adj Close'].iloc[:]
adj_close = adj_close.T.reset_index().iloc[:,[0,1,43]]
adj_close.columns = ['symbol', 'start_adj_close', 'end_adj_close']
adj_close['delta_adj_close'] = adj_close['end_adj_close'] - adj_close['start_adj_close']
# adj_close

volume = tickers_hist['Volume'].iloc[:]
volume = volume.T.reset_index().iloc[:,[0,1,43]]
volume.columns = ['symbol', 'start_vol', 'end_vol']
volume['delta_vol'] = volume['end_vol'] - volume['start_vol']
# volume

dividends = tickers_hist['Dividends'].iloc[:]
dividends = dividends.T.reset_index()
div1 = dividends.iloc[:,0]
div2 = dividends.iloc[:,1:43].apply(np.count_nonzero, axis=1)
div = pd.concat([div1, div2], axis=1)
div.columns = ['symbol', 'div_count']
div['div_action'] = np.where(div['div_count'] > 0, 1, 0) 
# div

splits = tickers_hist['Stock Splits'].iloc[:]
splits = splits.T.reset_index()
sp1 = splits.iloc[:,0]
sp2 = splits.iloc[:,1:43].apply(np.count_nonzero, axis=1)
sp = pd.concat([sp1, sp2], axis=1)
sp.columns = ['symbol', 'split_count']
sp['split_action'] = np.where(sp['split_count'] > 0, 1, 0) 
# sp

adj_close
```

### It will take about 1 hour to completely run through all market caps

```python
# Retrieving Market Cap from YFinance
def get_market_caps(tickers):
    market_caps = {}
    for ticker in tickers:
        try:
            stock = yf.Ticker(ticker)
            info = stock.info
            market_cap = info.get("marketCap")
            if market_cap is not None:
                market_caps[ticker] = market_cap
            else:
                market_caps[ticker] = None
        except Exception as e:
            print(f"Error getting data for {ticker}: {e}")
            market_caps[ticker] = None
        time.sleep(1)
    df = pd.DataFrame(list(market_caps.items()), columns=['Ticker', 'Market Cap'])
    return df
```

```python
stock_tickers = r3000_tickers
market_cap_df = get_market_caps(stock_tickers)

if market_cap_df is not None:
    print(market_cap_df)
```

### Classify the Cap Size based on Market Cap (previous cell)

```python
def classify_cap_size(adj_market_cap):
    if adj_market_cap >= 10_000_000_000:
        return 'Large Cap'
    elif 2_000_000_000 <= adj_market_cap < 10_000_000_000:
        return 'Medium Cap'
    elif adj_market_cap < 2_000_000_000:
        return 'Small Cap'
    else:
        return None
```

```python
market_cap_df = pd.read_csv('r3000_market_cap.csv')
market_cap_df.columns = ['symbol','market_cap']
market_cap_df['cap_size'] = market_cap_df['market_cap'].apply(classify_cap_size)
market_cap_df
```

### Final Dataset

```python
# Merge all datasets

merged_df = pd.merge(r3000_df, emp_df, on='symbol')
merged_df = pd.merge(merged_df, adj_close, on='symbol')
merged_df = pd.merge(merged_df, volume, on='symbol')
merged_df = pd.merge(merged_df, div, on='symbol')
merged_df = pd.merge(merged_df, sp, on='symbol')
merged_df = pd.merge(merged_df, market_cap_df, on='symbol')
merged_df
```
